# Tri par insertion

### Invariant de boucle: 

Notre sous tableau t de j a j-1 sera toujours trié.

- Initialisation : Un seul element dans notre tableau donc forcement trié
- Conservation : Si t de taille j est trié alors apres une itération, t de taille j+1 est toujours trié après insertion de A[j+1]
- Terminaison : Lorsque j = taille(A), t est trié et contient tout A (permutation)

### complexité théorique spatiale:

La taille de stockage de l'algorithme est fixe, il n'y a pas de relation directe avec la taille des intrants, Complexité spatiale est donc de O(1) car un seul element à stocker.
Une seul cas, Key sera la seul variable a stocker

### Complexité théorique temporel

Complexité de O(n2) car l'algo parcours deux boucles durant sont execution, celon la donnée entrée, tableau déjà trié, parsiellement trié, ou inversé, cette complexité est amenée à changer comme ci-dessous: 

- meilleur: le tableau est déjà trié donc compléxite de n car pas de passage dans le while
- moyen: A moitié trié, ne passe pas i fois dans la boucle
- pire: Le tableau est à l'envers, passage i fois dans la boucle

Cette algo va parcourir le tableau en testant si la valeur de i+1 superieur ou egale et stocker la valeur a interchanger dans une variable temporaire pour pouvoir switcher

```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 100
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 100))


def insertionsort(t):
    for i in range(1, len(t)):
        key = t[i]
        j = i-1
        while j >= 0 and key < t[j]:
            t[j+1] = t[j]
            j -= 1
            t[j+1] = key


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "insertionsort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "insertionsort(t_sorted)", number=num_runs, globals=globals())
duration_reverse = timeit.timeit(
    "insertionsort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()
```

# Tri par selection

### Invariant de boucle: 
A la fin de l'étape i, le tableau est une permutation du tableau initial et les i premiers éléments du tableau coïncident avec les i premiers éléments du tableau trié.

- Initialisation : Le premier élement de notre sous tableau j sera toujours trié comme pour le tri par insertion
- Conservation : Le sous tableau j sera toujours trié (main gauche) egalement comme pour le tri par insertion
- Terminaison : Le sous tableau j sera trié

### complexité théorique spatiale:

La taille de stockage de l'algorithme est fixe, il n'y a pas de relation directe avec la taille des intrants, Complexité spatiale est donc de O(1) car un seul element à stocker.
Un seul cas, l'index min est stocké en tant que variable temporaire pour pouvoir switcher les valeurs.

### Complexité théorique temporel

Complexité de Θ(n2) car l'algo parcours deux boucles durant sont execution, quelque soit le tableau d'entrée, le passage dans la deuxieme boucle est obligatoire contrairement au trie par insertion

meilleur, moyen, pire : Il n'y à pas de scenarios avantageux, les comparaisont et passage dans le boucle sont systematiques, il n'y a donc pas de gains de temps en fonction du tableau d'entré.

```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 100
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 100))


def selectionsort(t):
    for i in range(len(t)):
        min = i
        for j in range(i+1, len(t)):
            if t[min] > t[j]:
                min = j

        t[i], t[min] = t[min], t[i]


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "selectionsort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "selectionsort(t_sorted)", number=num_runs, globals=globals())
duration_reverse = timeit.timeit(
    "selectionsort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()
```

# Tri par Bulles

Le tri à bulles compare et permute à plusieurs reprises (si nécessaire) les éléments adjacents à chaque passage. Dans la i-ème passe de Bubble Tri (ordre croissant), les derniers (i-1) éléments sont déjà triés et le i-ème plus grand élément est placé à la (Ni)-ème position, c’est-à-dire la i-ème dernière position.

### Invariant de boucle: 

- Initialisation : Le premier élement de notre sous tableau j sera toujours trié comme pour le tri par insertion
- Conservation : Le sous tableau j sera toujours trié (main gauche) egalement comme pour le tri par insertion
- Terminaison : Le sous tableau j sera trié

### complexité théorique spatiale:

Une variable temporaire est utilisée dans l’échange [ auxiliaire, O(1) ]. Il s’agit donc d’un tri sur place. 

### Complexité théorique temporel
- Array trié dans le meilleur des cas en entrée. Ou presque tous les éléments sont à leur place. [ O(N) ]. O(1) s’échange, la complexité est linéaire.
- Moyenne : La complexité moyenne est de Θ(n2), le nombre de permutations de paires d'éléments successifs est égal au nombre d'inversions de la permutation
- Pire cas : trié à l’envers / Très peu d’éléments sont à la bonne place. [ O (N 2 ) ]. O(N 2 ), quand le plus petit élément est à la fin du tableau.

aide avec algo de geekforgeek: https://www.geeksforgeeks.org/


```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 100
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 100))


def bubblesort(t):
    n = len(t)
    for i in range(n):
        for j in range(0, n-i-1):
            if t[j] > t[j+1]:
                t[j], t[j+1] = t[j+1], t[j]


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "bubblesort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "bubblesort(t_sorted)", number=num_runs, globals=globals())
duration_reverse = timeit.timeit(
    "bubblesort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()
```

# Tri Fusion

### Complexité théorique spaciale :

- Meilleur : O(1) dans le meilleur
- Pire : O(n) dans le pire des cas

### Complexité théorique temporel :
- Pire des cas : O(n log n)
- Moyen des cas : O(n log n)
- Meilleur cas : O(n log n)

### Analyse théorique :

Algo diviser pour régner : 

On souhaite construire un tableau trié avec deux tableaux triés. Cela consiste donc à couper les tableaux en deux de longueur egale, trier ces derniers puis les combiner en un seul tableau trié.

### Comparaison avec implantation si dessous :

On peut voir dans les résultat si dessous que peut importe comment le tableau d'entré est trier, les temps de traitement sont identiques ce qui convient au fait que comme au divise notre tableau en plein de sous tableau de 2 voir 1 élément pour les trier la place de tous les éléments n'importe plus vraiment, comparer à un trie par insertion ou sélection.

source algo : https://www.geeksforgeeks.org/

```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 100
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 100))


def merge(left, right):
    if not len(left) or not len(right):
        return left or right

    result = []
    i, j = 0, 0
    while (len(result) < len(left) + len(right)):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
        if i == len(left) or j == len(right):
            result.extend(left[i:] or right[j:])
            break

    return result


def mergesort(t):
    if len(t) < 2:
        return t

    middle = int(len(t)/2)
    left = mergesort(t[:middle])
    right = mergesort(t[middle:])

    return merge(left, right)


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "mergesort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "mergesort(t_sorted)", number=num_runs, globals=globals())
duration_reverse = timeit.timeit(
    "mergesort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()
```

# Tri rapide

### Complexité spaciale
- Pire cas : o(n)
- Moyenne	: o(log n)

### Complexité en temporelle

- Pire cas : o(n2)
- Moyenne :  o(n log n)
- Meilleur cas : o(n)

### Analyse théorique :

Algo diviser pour régner : 

La méthode consiste à placer un élément du tableau (pivot) à sa place définitive, en permutant tous les éléments de telle sorte que tous ceux qui sont inférieurs au pivot soient à sa gauche et que tous ceux qui sont supérieurs au pivot soient à sa droite.

### Comparaison avec implantation si dessous :

On peut voir dans les résultat si dessous que peut importe comment le tableau d'entré est trier, les temps de traitement sont identiques.

```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 100
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 100))


def quicksort(t):
    inf, equal, sup = [], [], []
    if len(t) > 1:
        pivot = t[0]
        for i in t:
            if i < pivot:
                inf.append(i)
            if i == pivot:
                equal.append(i)
            if i > pivot:
                sup.append(i)
        return quicksort(inf) + equal + quicksort(sup)
    return t


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "quicksort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "quicksort(t_sorted)", number=num_runs, globals=globals())
duration_reverse = timeit.timeit("quicksort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()

```

# Tri Stupid


### Complexité spaciale

- Pire cas : o(n)

### Complexité en temporelle

- pire : infini
- moyen : o(n+1!)
- meilleur : o(n)

### Analyse théorique :

Le tri stupide consiste à vérifier si les éléments sont ordonnés et s'ils ne le sont pas, à les mélanger aléatoirement, puis à recommencer autant de fois que nécessaire. Cet algo est "probabiliste"

### Comparaison avec implantation si dessous :

On peut voir dans les résultat si que cette algo peu ne jamais se finir dans le pire des cas.

```python
import random
import timeit
import matplotlib.pyplot as plt
import numpy as np

num_runs = 10
t = []
t_reverse = []
t_sorted = []

x_rand = []
x_sorted = []
x_reverse = []

for k in range(num_runs):
    t.append(random.randint(0, 10))


def stupidsort(t):

    tri = True
    for i in range(len(t)-1):
        if t[i+1] < t[i]:
            tri = False
        while tri == False:
            i = random.randint(0, len(t)-1)
            j = i = random.randint(0, len(t)-1)
            t[i] = t[j]


t_reverse = sorted(t, reverse=True)
t_sorted = sorted(t)

duration_rand = timeit.timeit(
    "stupidsort(t)", number=num_runs, globals=globals())
duration_sorted = timeit.timeit(
    "stupidsort(t_sorted)", number=num_runs, globals=globals())
#duration_reverse = timeit.timeit(
    #"stupidsort(t_reverse)", number=num_runs, globals=globals())

x_rand.append(duration_rand)
x_sorted.append(duration_sorted)
#x_reverse.append(duration_reverse)

y = num_runs
plt.plot(y, x_rand, '-o', label="rand")
plt.plot(y, x_sorted, '-o', label="sorted")
#plt.plot(y, x_reverse, '-o', label="reverse")
plt.legend()
plt.show()
```
